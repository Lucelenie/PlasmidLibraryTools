This is a script that uses BioPython and uniprotenator to analyze our sequencing data.

```python
from Bio.Seq import Seq
import string

from urllib2 import urlopen
from lxml import objectify

from Bio import pairwise2
```

```python
plasmid_sequence_file = '../plasmid sequences/empty/pet_his6_nusa_tev_lic.seq'
filename = '../sequencing data by clone/5 - NTerm NusA/Nterm-NusA-T7-Term.seq'
```

```python
def rev_sequencing_results_no_translate(filename):
    with open(filename, 'r') as f:
        sequence_rough = f.readlines()[1:]
    sequence_rough = string.join(sequence_rough)
    nucleotide_sequence = sequence_rough.replace('\n ','').replace('\n','')
    seq = Seq(nucleotide_sequence)
    reverse_complement = seq.reverse_complement()
    
    return str(reverse_complement)
```

```python
sequencing_result = rev_sequencing_results_no_translate(filename)
```

```python
sequencing_result
```

How does this align with our plasmid sequence?

```python
with open(plasmid_sequence_file, 'r') as f:
    sequence_rough = f.readlines()[1:]
sequence_rough = string.join(sequence_rough)
plasmid_sequence = sequence_rough.replace('\n ','').replace('\n','')
    
```

```python
plasmid_sequence
```

```python
alignments = pairwise2.align.globalxx(sequencing_result, plasmid_sequence)
```
